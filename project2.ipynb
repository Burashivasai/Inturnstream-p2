```python
import pandas as pd

# Load dataset
df = pd.read_csv("9. Sales-Data-Analysis (2).csv")

# Display first 5 rows
print("Preview of dataset:")
print(df.head())

# Basic info about dataset
print("\nDataset Info:")
print(df.info())

# Shape of dataset
print("\nShape of dataset (rows, columns):")
print(df.shape)

# Summary statistics for numeric columns
print("\nSummary Statistics:")
print(df.describe())

# Count of missing values
print("\nMissing Values per Column:")
print(df.isnull().sum())

# Check for duplicates
print("\nNumber of duplicate rows:")
print(df.duplicated().sum())

# Categorical column summary (if any)
print("\nCategorical Value Counts:")
for col in df.select_dtypes(include='object').columns:
    print(f"\n{col}:")
    print(df[col].value_counts().head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("9. Sales-Data-Analysis (2).csv")

# Histogram for numeric columns
df.hist(figsize=(10, 6), bins=20, edgecolor="black")
plt.suptitle("Histograms of Numeric Columns")
plt.show()

# Box plots for numeric columns
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.select_dtypes(include='number'))
plt.title("Box Plots of Numeric Columns")
plt.xticks(rotation=45)
plt.show()

# Scatter plot between two numeric columns (example: col1 vs col2)
plt.figure(figsize=(8, 6))
sns.scatterplot(x="Price", y="Quantity", data=df)
plt.title("Scatter Plot: Price vs Quantity")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("9. Sales-Data-Analysis (2).csv")

# Correlation matrix (numeric columns only)
corr = df.corr(numeric_only=True)

print("Correlation Matrix:")
print(corr)

# Heatmap for correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()

# Pairplot to see patterns between variables
sns.pairplot(df.select_dtypes(include='number'))
plt.suptitle("Pairwise Relationships", y=1.02)
plt.show()

# Example: grouping pattern for categorical column (if present)
if not df.select_dtypes(include='object').empty:
    cat_col = df.select_dtypes(include='object').columns[0]  # first categorical column
    num_col = df.select_dtypes(include='number').columns[0]  # first numeric column

    plt.figure(figsize=(8, 6))
    sns.boxplot(x=cat_col, y=num_col, data=df)
    plt.title(f"Pattern of {num_col} across {cat_col}")
    plt.xticks(rotation=45)
    plt.show()
```
